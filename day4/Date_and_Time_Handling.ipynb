###Working with the Dates

```python
#Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

####1. Extract Year

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

####2. Extract Month

```python
date['date_month'] = date['date'].dt.month # but this gives the month number
date['date_month_name'] = date['date'].dt.month_name() # this gives the month name
date.sample(5)
```

####3. Extract Days

-Day number

```python
date['date_day'] = date['date'].dt.day
date.sample(5)
```

-Day of the week

```python
date['date_dow'] = date['date'].dt.dayofweek #Again this only provides wth the da of the wek as a number from 0 to 7
date['date_dow_name'] = date['date'].dt.day_name() #here to get name of the day we dont need to specify the dayofweek but general as day_name
date.sample(5)
```

-Weekend?

```python
#to check if the day is weekend we need to use logic and not inbuilt function
date['date_weekend?'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

####4. Extract Week of the Year

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

####5. Extract Quater

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

####6. Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1,2)
date.head()
```

####7. Extract Time elapsed between dates

```python
import datetime
today = datetime.datetime.today()
today
```

-Time passed

```python
today - date['date']
```

-Days passed exactly

```python
(today - date['date']).dt.days
```

-Months passed

```python
#This stuff doesn't come up at the door so need a little tweaks
#np.round((today - date['date']) / np.timedelta64(1, 'M'),0) ---> this method is discarded from the video
#therefore make a jugaad
np.round((today - date['date']).dt.days / 30.4375, 0)
```
